```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(['seaborn-notebook', 'seaborn-white'])

import mlgrad.func as func
```

```python
a=0.5
sq = func.Sqrt(a)
def h(z,f):
    return 0.5*(z+u - f(z-u))
```

```python
u=2.0
plt.figure(figsize=(6,5))
zs = np.linspace(-0,6,500)
plt.plot([-0,u], [0, u], color='k', linestyle='--', label=r'$\min(z,u)$')
plt.plot([u,3*u], [u, u], color='k', linestyle='--')
plt.plot(zs, [h(z, sq) for z in zs], color='k', label=r'$\frac{1}{2}(z+u-\sqrt{\varepsilon^2+(z-u)^2}+\varepsilon)$')
plt.ylim(0, 2.1)
plt.xlim(0, 6.1)
plt.xticks([0,u], ['0', 'u'])
plt.yticks([0,u], ['0', 'u'])
plt.vlines(u, 0, u, linestyles=':', linewidth=1.)
plt.hlines(u, 0, u, linestyles=':', linewidth=1.)
plt.legend(loc='lower right')
plt.savefig('fig/smoothed_cut.eps')
plt.show()
```
