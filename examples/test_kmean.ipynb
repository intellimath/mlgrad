```python
import numpy as np
from mlgrad.distance import PowerDistance, LogDistance
from mlgrad.kmean import KMean

import matplotlib.pyplot as plt
print(plt.style.available)
plt.style.use('bmh')
%matplotlib inline
```

```python
X1 = np.random.random((50,))
Y1 = np.random.random((50,))
X2 = np.random.random((50,)) + 0.75
Y2 = np.random.random((50,)) + 0.75
```

```python
plt.scatter(X1, Y1, c='r', s=36, label='1')
plt.scatter(X2, Y2, c='b', s=36, label='2')
plt.legend(loc='best')
plt.grid(1)
plt.minorticks_on()
```

```python
dst = LogDistance(2.0, 2)
```

```python
U1 = np.c_[X1, Y1]
U2 = np.c_[X2, Y2]
U = np.vstack((U1,U2))
print(U1.shape, U2.shape, U.shape)
```

```python
alg = KMean(dst, h=10.0, max_epoch=1000, eps=1.e-9)
```

```python
#alg.fit(U, centers=[[0.5,0.5],[1.5,1.5]])
alg.fit(U, n_centers=2)
```

```python
print(alg.K, alg.centers)
print(alg.classes)
```

```python
plt.scatter(U[:,0], U[:,1], c=alg.classes, s=36)
plt.scatter(alg.centers[:,0], alg.centers[:,1], s=156)
plt.grid(1)
plt.minorticks_on()
```
