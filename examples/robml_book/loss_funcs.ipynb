```python
import numpy as np
import matplotlib.pyplot as plt

import mlgrad.funcs as funcs
```

```python
abs_func = funcs.Absolute()
square_func = funcs.Square()
huber_func = funcs.Huber()
tukey_func = funcs.Tukey(np.sqrt(6.))
sqrt_func = funcs.Sqrt(1.0)
fsqrt_func = funcs.FSqrt(1.0, 0.25)
logsquare_func = funcs.LogSquare(1.0)
trunc_func = funcs.TruncAbs(1.)
gauss_func = funcs.Gauss(1.)
```

```python
range_x = np.array(np.linspace(-4, 4, 200), 'd')
range_x2 = np.array(np.linspace(-8, 8, 200), 'd')
range_x3 = np.array(np.linspace(-6, 6, 200), 'd')
plt.figure(figsize=(14,5))
plt.subplot(1,3,1)
plt.title(r'$\varrho(x) \to +\infty$, $|\varrho^\prime(x)| < C$')
plt.plot(range_x, abs_func.evaluate_array(range_x), label=r'$|x|$')
plt.plot(range_x, sqrt_func.evaluate_array(range_x), label=r'$\sqrt{1+x^2}-1$')
plt.plot(range_x, trunc_func.evaluate_array(range_x), label=r'$\max(0,|x|-c)$')
# plt.plot(range_x, huber_func.evaluate_array(range_x), label=r'huber')
plt.legend()
plt.minorticks_on()
# plt.ylim(-0.25, 4)

plt.subplot(1,3,2)
plt.title(r'$\varrho(x) \to +\infty$, $\varrho^\prime(x) \to 0$')
plt.plot(range_x2, logsquare_func.evaluate_array(range_x2), label=r'$\ln(1+x^2)$')
plt.plot(range_x2, fsqrt_func.evaluate_array(range_x2), label=r'$(1+x^2)^{1/4}$')

plt.legend()
plt.minorticks_on()

plt.subplot(1,3,3)
plt.title(r'$|\varrho(x)| < C$, $\varrho^\prime(x) \to 0$')
plt.plot(range_x2, tukey_func.evaluate_array(range_x2), label=r'tukey')
plt.plot(range_x2, np.ones_like(range_x2)-(gauss_func.evaluate_array(range_x2)), label=r'$1-e^{-x^2/2}$')
# plt.hlines(0, 0, 3.0, linewidth=0.5)
# plt.vlines(0, 0, 4.0, linewidth=0.5)
plt.legend()
plt.minorticks_on()
plt.tight_layout()
plt.show()
```

```python
range_x = np.asarray(np.linspace(-4, 4, 200), 'd')
range_x2 = np.asarray(np.linspace(-8, 8, 200), 'd')
range_x3 = np.asarray(np.linspace(-6, 6, 200), 'd')
plt.figure(figsize=(14,5))
plt.subplot(1,3,1)
plt.title(r'$\varrho(x) \to +\infty$, $|\varrho^\prime(x)| < C$')
# plt.plot(range_x2, square_func.derivative_div_array(range_x2), label=r'$\frac{1}{2}x^2$')
plt.plot(range_x2, sqrt_func.derivative_div_array(range_x2), label=r'$\sqrt{1+x^2}-1$')
plt.plot(range_x2, trunc_func.derivative_div_array(range_x2), label=r'$\max(0,|x|-1)$')
# plt.plot(range_x2, huber_func.evaluate_array(range_x2), label=r'huber')
plt.legend()
plt.minorticks_on()
# plt.ylim(-0.25, 4)

plt.subplot(1,3,2)
plt.title(r'$\varrho(x) \to +\infty$, $\varrho^\prime(x) \to 0$')
plt.plot(range_x2, logsquare_func.derivative_div_array(range_x2), label=r'$1/(1+\frac{1}{2}x^2)$')
plt.plot(range_x2, fsqrt_func.derivative_div_array(range_x2), label=r'$\frac{1}{4}x(1+x^2)^{-3/4}$')

plt.legend()
plt.minorticks_on()

plt.subplot(1,3,3)
plt.title(r'$|\varrho(x)| < C$, $\varrho^\prime(x) \to 0$')
plt.plot(range_x2, tukey_func.derivative_div_array(range_x2), label=r'tukey')
plt.plot(range_x2, -np.asarray(gauss_func.derivative_div_array(range_x2)), label=r'$e^{-x^2/2}$')
# plt.hlines(0, 0, 3.0, linewidth=0.5)
# plt.vlines(0, 0, 4.0, linewidth=0.5)
plt.legend()
plt.minorticks_on()
plt.tight_layout()
plt.show()
```
