```python
import numpy as np
import matplotlib.pyplot as plt

import sklearn.datasets as datasets
import sklearn.decomposition as decomposition
```

```python
X, Y = datasets.load_digits(return_X_y=True)
X /= 16
```

```python
I3 = (Y==3)
X3 = X[I3]
Y3 = Y[I3]
```

```python
n = X.shape[1]
pca = decomposition.PCA()
U = pca.fit_transform(X3)
```

```python
plt.bar(range(n), pca.explained_variance_ratio_)
plt.minorticks_on()
plt.show()
```

```python
plt.figure()
cb = plt.scatter(U[:,0], U[:,1], c=Y3, edgecolors='k', linewidths=0.5, cmap=plt.cm.gist_rainbow)
plt.colorbar(cb)
fig = plt.gcf()
fig.set_figheight(6)
fig.set_figwidth(10)
plt.show()
```
