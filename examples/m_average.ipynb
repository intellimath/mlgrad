```python
import mlgrad.func as func
import mlgrad.avragg as avragg
import matplotlib.pyplot as plt
import numpy as np

from mlgrad import averager_it, average_it
```

```python
X = 10*np.random.random((1000,))
X = X.astype('f')
```

```python
alphas = np.linspace(0.01, 0.99, 21)
ks = []
vs = []
for alpha in alphas:
    print(alpha)
    agg = averager_it(func.Quantile_Sqrt(alpha, 0.001))
    agg.fit(X)
    agg.nabla(X)
    ks.append(agg.K)
    vs.append(agg.u)
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1, 2, 1)
plt.plot(alphas, vs, label='u')
plt.legend()
plt.minorticks_on()
plt.grid(1)
plt.subplot(1, 2, 2)
plt.plot(alphas, ks, label='K')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
alphas = np.linspace(0.01, 0.99, 20)
ks = []
vs = []
for alpha in alphas:
    agg = avragg.WMAverage(averager_it(func.Quantile_Sqrt(alpha, 0.001)))
    agg.fit(X)
    ks.append(agg.K)
    vs.append(agg.u)
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1, 2, 1)
plt.plot(alphas, vs, label='u')
plt.legend()
plt.minorticks_on()
plt.grid(1)
plt.subplot(1, 2, 2)
plt.plot(alphas, ks, label='K')
plt.legend()
plt.tight_layout()
plt.show()
```
