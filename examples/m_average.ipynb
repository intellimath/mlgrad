```python
import mlgrad.func as func
import mlgrad.avragg as avragg
import matplotlib.pyplot as plt
import numpy as np

from mlgrad.af import averaging_function, scaling_function
```

```python
X = 10*np.random.random((1000,))
```

```python
alphas = np.linspace(0.01, 0.99, 21)
ks = []
vs = []
for alpha in alphas:
    # print(alpha)
    agg = averaging_function("M", func.quantile_func(alpha, func.Sqrt(0.001)), tol=1.0e-4)
    agg.fit(X)
    agg.gradient(X)
    ks.append(agg.K)
    vs.append(agg.u)
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1, 2, 1)
plt.plot(alphas, vs, label='u')
plt.legend()
plt.minorticks_on()
plt.grid(1)
plt.subplot(1, 2, 2)
plt.plot(alphas, ks, label='K')
plt.legend()
plt.tight_layout()
plt.show()
```
