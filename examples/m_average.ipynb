```python
import mlgrad.func as func
import mlgrad.avragg as avragg
import matplotlib.pyplot as plt
import numpy as np

from mlgrad import averager_it, average_it
```

```python
X = np.random.random((1000,))
```

```python
alphas = np.linspace(0.01, 5.00, 500)
ks = []
vs = []
for alpha in alphas:
    agg = average_it(X, func.ArithMod(alpha))
    ks.append(agg.K)
    vs.append(agg.u)
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1, 2, 1)
plt.plot(alphas, vs, label='u')
plt.legend()
plt.minorticks_on()
plt.grid(1)
plt.subplot(1, 2, 2)
plt.plot(alphas, ks, label='K')
plt.legend()
plt.tight_layout()
plt.show()
```
