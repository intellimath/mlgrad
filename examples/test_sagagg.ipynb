```python
import numpy as np
from mlgrad.avragg import Average_GD, Average_SAG
from mlgrad.func import Quantile, Expectile, Quantile_AlphaLog
import matplotlib.pyplot as plt
plt.style.use('bmh')

%matplotlib inline
%config BackendInline.format='svg'
```

```python
#X = 1.0 + 0.1*np.random.random((1000,))
X = np.random.random((1000,))
```

```python
vs = []
ks = []
hs = []
alphas = np.linspace(0.05, 0.95, 200)
m = len(alphas)
for alpha in alphas:
    dissim = Quantile(alpha)
    alg = Average_SAG(dissim, h=0.001, tol=1.0e-5)
    alg.fit(X)
    vs.append(alg.y)
    ks.append(alg.K)    
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1, 3, 1)
plt.plot(alphas, vs)
plt.xlim(0., 1.)
#plt.ylim(0., 1.)
plt.subplot(1, 3, 2)
plt.plot(alphas, ks)
# plt.subplot(1, 3, 3)
# plt.plot(alphas, hs)
# plt.tight_layout()
plt.show()
```

```python
vs = []
ks = []
hs = []
alphas = np.linspace(0.05, 0.95, 200)
m = len(alphas)
for alpha in alphas:
    penalty = Expectile(alpha)
    alg = Average_SAG(penalty, h=0.01, tol=1.0e-5)
    alg.fit(X)
    vs.append(alg.y)
    ks.append(alg.K)
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1, 3, 1)
plt.plot(alphas, vs)
plt.subplot(1, 3, 2)
plt.plot(alphas, ks)
# plt.subplot(1, 3, 3)
# plt.plot(alphas, hs)
plt.tight_layout()
plt.show()
```

```python
vs = []
ks = []
hs = []
alphas = np.linspace(0.1, 100., 100)
m = len(alphas)
for alpha in alphas:
    penalty = Quantile_AlphaLog(alpha)
    alg = Average_SAG(penalty, h=0.01)
    alg.fit(X)
    vs.append(alg.y)
    ks.append(alg.K)
    hs.append(alg.h_mean)
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1, 3, 1)
plt.plot(alphas, vs)
plt.subplot(1, 3, 2)
plt.plot(alphas, ks)
plt.subplot(1, 3, 3)
plt.plot(alphas, hs)
plt.tight_layout()
plt.show()
```
