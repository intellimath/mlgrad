```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn.preprocessing as preprocessing
```

```python
df = pd.read_csv("data/radarimage.csv", index_col="rownames")
df
```

```python
X = df["X.coord"].values
Y = df["Y.coord"].values
B1 = df["Band.1"].values
B2 = df["Band.2"].values
B3 = df["Band.3"].values
```

```python
B1 = preprocessing.robust_scale(B1)
B2 = preprocessing.robust_scale(B2)
B3 = preprocessing.robust_scale(B3)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(X, B1, s=9, c='w', edgecolors='r', label='B1')
plt.scatter(X, B2, s=9, c='w', edgecolors='g', label='B2')
plt.scatter(X, B3, s=9, c='w', edgecolors='b', label='B3')
plt.legend()
plt.subplot(1,2,2)
plt.scatter(Y, B1, s=9, c='w', edgecolors='r', label='B1')
plt.scatter(Y, B2, s=9, c='w', edgecolors='g', label='B2')
plt.scatter(Y, B3, s=9, c='w', edgecolors='b', label='B3')
plt.legend()
plt.show()
```
