```python
import numpy as np
import mlgrad.avragg as avragg
import mlgrad.averager as averager
import mlgrad.func as func
import matplotlib.pyplot as plt
plt.style.use(['seaborn-paper', 'seaborn-white'])
```

```python
X = np.random.random((1000,))
```

```python
vs = []
ks = []
alphas = np.linspace(0.05, 0.95, 200)
m = len(alphas)
for alpha in alphas:
    pfunc = func.Quantile(alpha)
    #alg = avragg.Average_FG(pfunc)
    alg = avragg.average_fg(X, pfunc)
    vs.append(alg.u)
    ks.append(alg.K)
```

```python
plt.figure(figsize=(12,3))
plt.subplot(1, 3, 1)
plt.plot(alphas, vs)
plt.xlim(0., 1.)
#plt.ylim(0., 1.)
plt.subplot(1, 3, 2)
plt.plot(alphas, ks)
plt.tight_layout()
plt.show()
```

```python
vs = []
ks = []
alphas = np.linspace(0.05, 0.95, 200)
m = len(alphas)
for alpha in alphas:
    penalty = func.Expectile(alpha)
    #alg = avragg.Average_FG(penalty, h=0.01, tol=0.00001)
    alg = avragg.average_fg(X, penalty, verbose=0)
    vs.append(alg.u)
    ks.append(alg.K)
```

```python
plt.figure(figsize=(12,3))
plt.subplot(1, 3, 1)
plt.plot(alphas, vs)
plt.subplot(1, 3, 2)
plt.plot(alphas, ks)
plt.tight_layout()
plt.show()
```

```python
vs = []
ks = []
alphas = np.linspace(0.001, 0.01, 100)
m = len(alphas)
for alpha in alphas:
    penalty = func.Quantile_AlphaLog(alpha, 0.5)
    alg = avragg.average_fg(X, penalty)
    vs.append(alg.u)
    ks.append(alg.K)
```
