```python
import numpy as np
import matplotlib.pyplot as plt
import mlgrad.af as af
import mlgrad.funcs as funcs
import mlgrad.utils as utils
import mlgrad.pca as pca
import mlgrad.plots as plots
import pandas as pd
import sklearn.datasets as datasets
```

```python
X, Y = datasets.load_iris(return_X_y=True)
```

```python
X0 = X
Y0 = Y
c = pca.location(X0)
print(c)
```

```python
As,Ls = pca.find_pc_all(X0-c, 3, verbose=True, return_us=False)
print(Ls)
print(As)
Us = (X0 - c) @ As.T
print(Us)
```

```python
plt.plot(Ls, marker='o')
plt.show()
```

```python
plt.scatter(Us[:,0], Us[:,1], c=Y0, edgecolors='k', s=25)
plt.show()
```

```python
c1 = pca.location_l1(X0, verbose=True)
```

```python
Ls1, As1, Us1 = pca.find_pc_l1_all(X0-c1, 3, verbose=True)
print(Ls1)
print(As1)
```
