# Принцип минимизации эмпирического риска

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(['seaborn-poster'])
```

# Среднее арифметическое

```python
X = np.random.random(size=17)
range_x = np.asarray(range(1,18))
plt.figure(figsize=(12,6))
plt.plot(range_x, 7+2*X-1, marker='o', label='данные')
plt.hlines(7, 1, 18)
plt.xticks(range_x)
plt.legend()
plt.show()
```
