```python
import numpy as np
import mlgrad.func as func
from mlgrad import average_it

import matplotlib.pyplot as plt
```

```python
X = np.random.random((100,))
```

```python
vs = []
ks = []
alphas = np.linspace(0.05, 0.95, 50)
m = len(alphas)
for alpha in alphas:
    X = np.ones(1000, 'd')
    M = int(100 * alpha)-1
    X[-M:] = 100
    f = func.Quantile_Sqrt(0.001, alpha)
    alg = average_it(X, f)
    ks.append(alg.K)
    vs.append(alg.u)
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1, 2, 1)
plt.plot(alphas, vs, label='u')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(alphas, ks, label='K')
plt.legend()
plt.tight_layout()
plt.show()
```
