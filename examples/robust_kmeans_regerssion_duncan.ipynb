```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib as mpl
mpl.style.use('seaborn-notebook')
```

```python
df_duncan = pd.read_csv("data/Duncan.csv", sep=",")
df_duncan
```

```python
X_name, X_type, X_income, X_education, X_prestige = \
           [df_duncan[name].values for name in ('name','type','income', 'education', 'prestige')]
```

```python
X_income = X_income.astype('d')
X_prestige = X_prestige.astype('d')
X_education = X_education.astype('d')
```

```python
import mlgrad
import mlgrad.model as ml_model
import mlgrad.risk as ml_risk
import mlgrad.loss as ml_loss
import mlgrad.weights as ml_weights
import mlgrad.func as ml_func
import mlgrad.af as ml_af
```

```python
layer = ml_model.GeneralModelLayer(1)
tone_linmod1 = ml_model.LinearModel(1)
tone_linmod2 = ml_model.LinearModel(1)
layer.append(tone_linmod1)
layer.append(tone_linmod2)
nn = ml_model.FFNetworkModel()
nn.add(layer)
# for mod in layer.models:
#     print(mod.n_param)
# print(layer.n_param)
# print(nn.n_param)

nn.allocate()
# print(np.array(nn.param))
# for mod in layer:
#     print(np.array(mod.param))

# print('*')
nn.init_param()

# print(np.array(nn.param))
# for mod in layer:
#     print(np.array(mod.param))
```

```python
layer = nn.layers[0]
layer.param.base
```

```python
minloss = ml_loss.SoftMinLoss2(ml_loss.ErrorLoss(ml_func.SoftAbs(0.001)), 2, a=15)
tone_risk = ml_risk.ER21(X_income.reshape(-1,1), X_prestige, nn, minloss)
```

```python
tone_alg = mlgrad.erm_fg(tone_risk, n_restart=5, h=0.1)

plt.plot(tone_alg.lvals)
plt.show()
```

```python
print(np.array(tone_linmod1.param))
print(np.array(tone_linmod2.param))
tone_alg.K
```

```python
plt.figure(figsize=(7,5))
# plt.figure(figsize=(15,12))
plt.scatter(X_income, X_prestige, s=49, c='w', edgecolors='k', linewidth=1.0)
# plt.title("Соотношения тонов")
plt.plot([0., 100.], tone_linmod1.evaluate_all(np.array([[0.], [100.]])), linestyle='--', label='1', color='k')
plt.plot([0., 100.], tone_linmod2.evaluate_all(np.array([[0.], [100.]])), linestyle=':', label='2', color='k')
# for x_name, x_income, x_prestige in zip(X_name, X_income, X_prestige):
#     plt.text(x_income, x_prestige+2, x_name)
# plt.xlabel('Фактическое соотношение тонов')
# plt.ylabel('Воспринимаемое соотношение тонов')
# plt.ylim(1.2, 3.1)
# plt.xlim(1.2, 3.1)
plt.legend()
# plt.savefig('fig/tonedata_scatter.eps')
plt.show()
```

```python
layer_rob = ml_model.GeneralModelLayer(1)
tone_rob_linmod1 = ml_model.LinearModel(1)
tone_rob_linmod2 = ml_model.LinearModel(1)
layer_rob.append(tone_rob_linmod1)
layer_rob.append(tone_rob_linmod2)
nn_rob = ml_model.FFNetworkModel()
nn_rob.add(layer_rob)
# for mod in layer.models:
#     print(mod.n_param)
# print(layer.n_param)
# print(nn.n_param)

nn_rob.allocate()
# print(np.array(nn.param))
# for mod in layer:
#     print(np.array(mod.param))

# print('*')
nn_rob.init_param()

# print(np.array(nn.param))
# for mod in layer:
#     print(np.array(mod.param))
```

```python
minloss = ml_loss.SoftMinLoss2(ml_loss.ErrorLoss(ml_func.SoftAbs(0.001)), 2, a=15)
tone_rob_risk = ml_risk.ER21(X_income.reshape(-1,1), X_prestige, nn_rob, minloss)
agg_func = ml_af.averaging_function("WM", alpha=0.9)
weights_tone = ml_weights.MWeights(agg_func, tone_rob_risk)
```

```python
fg_tone = mlgrad.fg(tone_rob_risk, h=0.1, tol=1.0e-9)
tone_rob_alg = mlgrad.erm_irgd(fg_tone, weights_tone)

plt.plot(tone_rob_alg.lvals)
plt.show()
```

```python
plt.figure(figsize=(7,5))
# plt.figure(figsize=(15,12))
plt.scatter(X_income, X_prestige, s=49, c='w', edgecolors='k', linewidth=1.0)
# plt.title("Соотношения тонов")
plt.plot([0., 100.], tone_rob_linmod1.evaluate_all(np.array([[0.], [100.]])), linestyle='--', label='1', color='k')
plt.plot([0., 100.], tone_rob_linmod2.evaluate_all(np.array([[0.], [100.]])), linestyle=':', label='2', color='k')
plt.plot([0., 100.], tone_linmod1.evaluate_all(np.array([[0.], [100.]])), linestyle='--', label='1', color='m')
plt.plot([0., 100.], tone_linmod2.evaluate_all(np.array([[0.], [100.]])), linestyle=':', label='2', color='m')
# for x_name, x_income, x_prestige in zip(X_name, X_income, X_prestige):
#     plt.text(x_income, x_prestige+2, x_name)
# plt.xlabel('Фактическое соотношение тонов')
# plt.ylabel('Воспринимаемое соотношение тонов')
# plt.ylim(1.2, 3.1)
# plt.xlim(1.2, 3.1)
plt.legend()
# plt.savefig('fig/tonedata_scatter.eps')
plt.show()
```
