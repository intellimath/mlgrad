```python
import numpy as np
import matplotlib.pyplot as plt

import sklearn.datasets as datasets
import sklearn.decomposition as decomposition
```

```python
X, Y = datasets.fetch_olivetti_faces(return_X_y=True)
```

```python
pca0 = decomposition.PCA()
X = pca0.fit_transform(X)
plt.bar(range(3), pca0.explained_variance_ratio_[:3])
plt.minorticks_on()
plt.show()
```

```python
U = pca0.fit_transform(X)
```

```python
cb = plt.scatter(U[:,0], U[:,1], edgecolors='k', linewidths=0.5, cmap=plt.cm.gist_rainbow)
for i in range(len(U)):
    plt.text(U[i,0], U[i,1], str(Y[i]))
plt.colorbar(cb)
fig = plt.gcf()
fig.set_figheight(6)
fig.set_figwidth(10)
plt.show()
```
