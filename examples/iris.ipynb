```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

import sklearn.decomposition as decomposition

import mlgrad as ml
```

```python
iris_df = pd.read_csv("data/iris.data", sep=',')
#iris_df
```

```python
X = np.c_[(iris_df['sl'], iris_df['sw'], iris_df['pl'], iris_df['pw'])]
cls_names = iris_df['cls'].unique().tolist()
Y = iris_df['cls'].apply(lambda x: cls_names.index(x)+1).values
```

```python
pca = decomposition.PCA()
U = pca.fit_transform(X)
```

```python
for i in range(4):
    plt.hist(U[:, i], align='left', label=str(i))
plt.legend()
plt.show()
```

```python
plt.scatter(U[:,0], U[:,1], c=Y)
plt.show()
```

```python
mask = U[:,0] > -1.5
U1 = U[mask]
Y1 = Y[mask]
```

```python
plt.scatter(U1[:,0], U1[:,1], c=Y1)
plt.show()
```

```python
pca1 = decomposition.PCA()
pca1.n_components=3
V = pca1.fit_transform(U1)

plt.bar(range(3), pca1.explained_variance_ratio_)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
k = 1
for i in range(2):
    for j in range(i+1, 3):
        plt.subplot(1, 3, k)
        plt.scatter(V[:,i], V[:,j], c=Y1)
        k += 1
plt.show()
```
