# Scattering spectra

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 2
N = 32
x0 = 8*np.random.random(M)
y0 = 8*np.random.random(M)

XE0 = 1*np.random.random(size=(N,M))-0.5
YE0 = 1*np.random.random(size=(N,M))-0.5

X = x0[None,:] + XE0
Y = y0[None,:] + YE0
```

```python
plt.scatter(X, Y)
plt.show()
```
